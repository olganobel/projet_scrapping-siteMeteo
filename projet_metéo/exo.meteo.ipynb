```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url='https://www.lachainemeteo.com/meteo-france/previsions-meteo-france-aujourdhui'
r = requests.get(url)
print(url, r.status_code)
soup = BeautifulSoup(r.content,'html')
```

```python
print (soup)
```

```python
listejours=[]

for element in soup.find_all('ol'):
   # print (element.text)
    for climat in element('li',attrs={'class':'day'}):
        print(climat('abbr')[0].text)
        print('############')
        listejours.append(climat('abbr')[0].text)
listejours 

len(listejours)

       
        

```

```python
listenumero=[]

for element in soup.find_all('ol'):
   # print (element.text)
    for climat in element('li',attrs={'class':'day'}):
        print(climat('div',attrs ={'class':'numDay'})[0].text)
        print('############')
        listenumero.append(climat('div',attrs ={'class':'numDay'})[0].text)
listenumero        

len(listenumero)
     
```

```python
listemois=[]

for element in soup.find_all('ol'):
   # print (element.text)
    for climat in element('li',attrs={'class':'day'}):
        print(climat('div',attrs ={'class':'nameMonth'})[0].text)
        print('############')
        listemois.append(climat('div',attrs ={'class':'nameMonth'})[0].text)
listemois 

len(listemois)
```

```python
liste=[]
for element in soup.find_all('ol'):
   # print (element.text)
    for climat in element('li',attrs={'class':'day'}):
        print(climat('a')[0].get('href'))
        liste.append(climat('a')[0].get('href'))
        print('#################')
```

```python
liste
```

```python
#automatisation de tous les liens
listedescription=[]
for element in liste:
    url= element 
    r = requests.get(url)
    print(url, r.status_code)
    soup = BeautifulSoup(r.content,'html')
    for meteo in soup.find_all('div',attrs={'id':'weatherSituation'}):
        print(meteo.text)
        listedescription.append(meteo.text)
len(listedescription)
    
```

```python
df=pd.DataFrame({'jour':listejours,'date':listenumero,'mois':listemois,'description':listedescription})
df
```

```python
df.to_csv('exo.meteo.csv',sep=',')
```
